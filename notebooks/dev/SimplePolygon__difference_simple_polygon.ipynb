# SimplePolygon__difference_simple_polygon

```python
from crossproduct import Point, SimplePolygon
import matplotlib.pyplot as plt
```

```python
def plot_difference(pg,pg1):
    if pg.nD==2:
        fig, ax = plt.subplots(figsize=(4,4))
    else:
        fig = plt.figure()
        ax = fig.add_subplot(111, projection='3d')
    pg.plot(ax,color='blue')
    pg.triangles.plot(ax,color='blue',alpha=0.1)
    pg1.plot(ax,color='orange')
    pg1.triangles.plot(ax,color='orange',alpha=0.1)
    pgs=pg.difference_simple_polygon(pg1)
    print('%s polygons' % (len(pgs)))
    pgs.plot(ax,color='lime')
    if pg.nD==2:
        for pg in pgs:
            x,y=[pt.x for pt in pg]+[pg[0].x],[pt.y for pt in pg]+[pg[0].y]
            ax.fill(x,y,color='lime',alpha=0.25)
    
    # sets x and y axis limits to the same values
    xlim,ylim=ax.get_xlim(),ax.get_ylim()
    min_lim, max_lim = min(xlim[0],ylim[0]), max(xlim[1],ylim[1])  
    ax.set_xlim(min_lim,max_lim), ax.set_ylim(min_lim,max_lim)
    if pg.nD==3:
        zlim=ax.get_zlim()
        min_lim, max_lim = min(xlim[0],ylim[0],zlim[0]), max(xlim[1],ylim[1],zlim[0])  
        ax.set_zlim(min_lim,max_lim)
    return fig, ax
```

```python
pg=SimplePolygon(Point(0.0,8.0,3.0),
                 Point(3.0,8.0,3.0),
                 Point(3.0,8.0,0.0),
                 Point(0.0,8.0,0.0))
pg1=SimplePolygon(Point(0.5,8.0,2.5),
                  Point(2.5,8.0,2.5),
                  Point(2.5,8.0,1.0),
                  Point(0.5,8.0,1.0))
fig,ax=plot_difference(pg,pg1)
```

```python
# no intersection
pg=SimplePolygon(Point(0,0),Point(1,0),Point(1,1),Point(0,1))
pg1=SimplePolygon(Point(0,2),Point(1,2),Point(1,3),Point(0,3))
fig,ax=plot_difference(pg,pg1)
```

```python
# point intersection
pg=SimplePolygon(Point(0,0),Point(1,0),Point(1,1),Point(0,1))
pg1=SimplePolygon(Point(1,1),Point(2,1),Point(2,2),Point(1,2))
fig,ax=plot_difference(pg,pg1)
```

```python
# full edge intersection
pg=SimplePolygon(Point(0,0),Point(1,0),Point(1,1),Point(0,1))
pg1=SimplePolygon(Point(0,1),Point(1,1),Point(1,2),Point(0,2))
fig,ax=plot_difference(pg,pg1)
```

```python
# half edge overlap intersection
pg=SimplePolygon(Point(0,0),Point(1,0),Point(1,1),Point(0,1))
pg1=SimplePolygon(Point(0.5,1),Point(1.5,1),Point(1.5,2),Point(0.5,2))
fig,ax=plot_difference(pg,pg1)
```

```python
# partial internal edge intersection
pg=SimplePolygon(Point(0,0),Point(1,0),Point(1,1),Point(0,1))
pg1=SimplePolygon(Point(0.25,1),Point(0.75,1),Point(0.75,2),Point(0.25,2))
fig,ax=plot_difference(pg,pg1)
```

```python
# full intersection
pg=SimplePolygon(Point(0,0),Point(1,0),Point(1,1),Point(0,1))
pg1=pg
fig,ax=plot_difference(pg,pg1)
```
