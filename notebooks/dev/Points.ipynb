```python
import matplotlib.pyplot as plt
from crossproduct import Point, Points
```

```python
#2D plot
fig, ax = plt.subplots()
Points(Point(0,0),Point(1,1)).plot(ax,'ro')
```

```python
#3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
Points(Point(0,0,0),Point(1,1,1)).plot(ax,'ro')
```
