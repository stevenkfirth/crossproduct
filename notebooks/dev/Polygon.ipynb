```python
import matplotlib.pyplot as plt
from crossproduct import Point, Polygon
```

```python
#2D
fig, ax = plt.subplots()
#ax.fill([0,1,1,0],[0,0,1,1])
Polygon(Point(0,0),Point(1,0),Point(1,1),Point(0,1)).plot(ax,alpha=0.5,
                                                          edgecolor='b')
Polygon(Point(0.5,0.5),Point(1.5,0.5),Point(1.5,1.5),Point(0.5,1.5)).plot(ax,alpha=0.5,color='tab:blue',
                                                                          edgecolor='b')
```

```python
#3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
Polygon(Point(0,0,0),Point(1,0,0),Point(1,1,1),Point(0,1,1)).plot(ax,alpha=0.5)
```

```python
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d') # Axes3D(fig)
x = [0,1,1,0]
y = [0,0,1,1]
z = [0,1,0,1]
verts = [list(zip(x,y,z))]
print(verts)
ax.add_collection3d(Poly3DCollection(verts))
```

```python
#3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
pg=Polygon(Point(0,0,0),Point(1,0,0),Point(1,1,1),Point(0,1,1))
verts=[[tuple(pt) for pt in pg]]
print(verts)
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
pc=Poly3DCollection(list(verts),alpha=0.5)
#pc.set_alpha(0.5)
ax.add_collection3d(pc)
```
