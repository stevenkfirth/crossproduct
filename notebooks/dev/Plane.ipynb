```python
import matplotlib.pyplot as plt
from crossproduct import Point, Vector, Plane
```

```python
#3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlim(-1,1), ax.set_ylim(-1,1), ax.set_zlim(-1,1)
Plane(Point(0,0,0),Vector(1,0,1)).plot(ax,alpha=0.5)
```

```python
# intersect_plane
pl1=Plane(Point(0,0,0),Vector(1,0,1))
pl2=Plane(Point(0,0,0),Vector(0,-1,1))
result=pl1.intersect_plane(pl2)
#
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlim(-1,1), ax.set_ylim(-1,1), ax.set_zlim(-1,1)
pl1.plot(ax,alpha=0.5), pl2.plot(ax,alpha=0.5,color='tab:orange'), result.plot(ax,color='r')
plt.show()
```
