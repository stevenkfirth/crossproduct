```python
import matplotlib.pyplot as plt
from crossproduct import Point, Vector
```

```python
#2D plot
fig, ax = plt.subplots()
Point(1,1).plot(ax,'ro')
```

```python
#3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
Point(1,1,1).plot(ax,'ro')
```

```python
# add
pt=Point(0,0)
v=Vector(1,1)
result=pt+v
fig, ax = plt.subplots()
pt.plot(ax,'bo'), v.plot(ax), result.plot(ax,'ro')
```

```python
# sub
pt1=Point(1,1)
pt2=Point(0,0)
result=pt1-pt2
fig, ax = plt.subplots()
pt1.plot(ax,'bo'), pt2.plot(ax,'bo'), result.plot(ax,color='r')
```
