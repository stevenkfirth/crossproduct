# SimplePolygon

```python
from crossproduct import Point, SimplePolygon
import matplotlib.pyplot as plt
#from mpl_toolkits.mplot3d import Axes3D
```

```python
def plot_intersection(pg,pg1):
    if pg.nD==2:
        fig, ax = plt.subplots(figsize=(4,4))
    else:
        fig = plt.figure()
        ax = fig.add_subplot(111, projection='3d')
    pg.plot(ax,color='blue')
    pg.triangles.plot(ax,color='blue',alpha=0.1)
    pg1.plot(ax,color='orange')
    pg1.triangles.plot(ax,color='orange',alpha=0.1)
    pts,pls,pgs=pg.intersect_simple_polygon(pg1)
    print('%s points; %s polylines; %s polygons' % (len(pts),len(pls),len(pgs)))
    pts.plot(ax,color='red',marker='o')
    pls.plot(ax,color='cyan')
    pgs.plot(ax,color='lime')
    
    # sets x and y axis limits to the same values
    xlim,ylim=ax.get_xlim(),ax.get_ylim()
    min_lim, max_lim = min(xlim[0],ylim[0]), max(xlim[1],ylim[1])  
    ax.set_xlim(min_lim,max_lim), ax.set_ylim(min_lim,max_lim)

    return fig, ax
```

## intersect_simple_polygon

```python
# internal intersection, 2 edges
pg=SimplePolygon(Point(0,0),Point(1,0),Point(1,1),Point(0,1))
pg1=SimplePolygon(Point(0.25,0),Point(0.75,0),Point(0.75,1),Point(0.25,1))
fig,ax=plot_intersection(pg,pg1)
```
